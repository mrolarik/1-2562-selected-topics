## การค้นหาด้วย Regular expression

### Regular Expressions

*  Regular Expressions are used in programming languages to filter texts or textstrings.  
It's possible to check, if a text or a string matches a regular expression  
*  Regular expression ใช้กันอย่างแพร่หลายในการเขียนโปรแกรมสำหรับการกรอง (filter) ข้อความ 

```
# การค้นหาแบบง่าย
s = "Regular expressions easily explained!"

print("easily" in s)
```

```
# การค้นหาแบบง่ายจะไม่สามารถค้นเจอหากใช้เครื่องหมายพิเศษ เช่น (* ? .)
# คำที่นำไปค้นหาจะต้องเหมือนกันเท่านั้น
print("Easil?" in s)
```

แสดงตัวอย่างการ import และใช้งาน Regular expression (re)

```
import re

x = re.search("easily","Regular expressions easily explained!")

print(x)
print(type(x))

# print value in the object
dir(x)
```

* ค้นหาคำว่า "easily" พบจากข้อความ "Regular expressions easily explained!"
* ซึ่งอยู่ในอินเด็กซ์ (index) ที่ 20-26

```
print(x)
print('string: ', x.string)
print('span: ', x.span())
print('start: ', x.start())
print('end: ', x.end())
```

```
import re

match = re.search("easily","Regular expressions easily explained!")

if match:
  print('Found:', match.group())
  print('At:', match.span())
else:
  pass 
```

### ทดสอบเมธอด (Method) **search** และ **findall**

```
import re

# search - หากค้นเจอคำที่ต้องการจะหยุดค้นทันที
x = re.search(r"ca","A cat and a rat can't be friends.")
print(x)

# findall - จะค้นคำที่ต้องการจนจบประโยค
x = re.findall(r"ca","A cat and a rat can't be friends.")
print(x)
```

```
import re

x = re.findall(r"ca","A cat and a rat can't be friends.")
print(x)

x = re.findall(r"rat","A cat and a rat can't be friends.")
print(x)
```

ใช้เงื่อนไขในการค้นหา **[rc]at** หมายถึง การค้นหา คำที่ขึ้นต้นด้วยตัวอักษร r หรือ c  ([rc]) และตามด้วย at ซึ่งจะเป็นไปได้ทั้ง cat และ rat

```
x = re.findall(r"[rc]at","A cat and a rat can't be friends.")
print(x)
```

ใช้เงื่อนไขในการค้นหา **[rc]a[nt] หมายถึง การค้นหา คำที่ขึ้นต้นด้วยตัวอักษร r หรือ c  ([rc]) และตามด้วยตัวอักษร a และลงท้ายด้วยตัวอักษร n หรือ t ([nt]) ซึ่งจะเป็นไปได้ทั้ง cat, can และ rat

```
x = re.findall(r"[cr]a[nt]","A cat and a rat can't be friends.")
print(x)
```

```
x = re.search(r"[cr]a[nt]","A cat and a rat can't be friends.")
print(x)
```

##  การค้นหาด้วย subprocess ด้วยคำสั่งของ linux 

### Python Execute Unix Command

คำสั่ง ls คือคำสั่งการเรียกดูข้อมูลที่อยู่ในไดเรกทอรี

```
! ls
```

```
!ls sample_data
```

* เครื่องหมาย | ในลินุกซ์จะเรียกว่าไปป์ (PIPE) เป็นการส่ง output ต่อไปยังคำสั่งถัดไป
* ดังนั้น output ของคำสั่ง ls 'sample_data' จะถูกส่งต่อไปยังคำสั่งที่สอง ซึ่งก็คือ grep 'mnist'
* คำสั่ง grep เป็นการค้นหาข้อความ ดังนั้น grep 'mnist' คือการค้นหาข้อความ 'mnist' ที่อยู่ใน output ที่ส่งมาจากคำสั่งแรก

```
!ls 'sample_data' | grep 'mnist'
```

### คำสั่ง subprocess ใช้สำหรับสั่งให้ linux ทำงานผ่าน command line

จากตัวอย่างของ subprocess ดังต่อไปนี้ เปรียบได้กับคำสั่ง linux ดังนี้  **$ls 'sample_data'**


```
import subprocess

out = subprocess.Popen(['ls', 'sample_data'], stdout=subprocess.PIPE, stderr=subprocess.STDOUT)
stdout,stderr = out.communicate()

print(stdout.decode('utf-8'))
```

จากตัวอย่างของ subprocess ดังต่อไปนี้ เปรียบได้กับคำสั่ง linux ดังนี้  **!ls 'sample_data' | grep 'mnist'**

```
import subprocess

proc1 = subprocess.Popen(['ls', 'sample_data'], stdout=subprocess.PIPE)
proc2 = subprocess.Popen(['grep', 'mnist'], stdin=proc1.stdout,
                        stdout=subprocess.PIPE, stderr=subprocess.PIPE)

proc1.stdout.close()
stdout,stderr = proc2.communicate()

print(stdout.decode('utf-8'))
```

สามารถใช้สัญลักษ์พิเศษในการค้นหาได้ เช่น (*)

```
import subprocess

proc1 = subprocess.Popen(['ls', 'sample_data'], stdout=subprocess.PIPE)
proc2 = subprocess.Popen(['grep', 'cali*'], stdin=proc1.stdout,
                        stdout=subprocess.PIPE, stderr=subprocess.PIPE)

proc1.stdout.close()
stdout,stderr = proc2.communicate()

print(stdout.decode('utf-8'))
```

### การใช้งานคำสั่ง curl ร่วมกับ subprocess

ใช้คำสั่ง curl เพื่ออ่านข้อมูลจากไฟล์ที่อยู่บนอินเทอร์เน็ต

```
!curl https://www.python-course.eu/simpsons_phone_book.txt
```

#### ใช้คำสั่ง crul ร่วมกับคำสั่ง grep ในกรณีที่ค้นหาแบบ case sensitive

```
import subprocess

search_txt = 'ot'
proc1 = subprocess.Popen(['curl', 'https://www.python-course.eu/simpsons_phone_book.txt'], stdout=subprocess.PIPE)
proc2 = subprocess.Popen(['grep', search_txt], stdin=proc1.stdout,
                        stdout=subprocess.PIPE, stderr=subprocess.PIPE)

stdout,stderr = proc2.communicate()

print(stdout.decode('utf-8'))
```

#### ใช้คำสั่ง crul ร่วมกับคำสั่ง grep ในกรณีที่ค้นหาแบบไม่เป็น case sensitive (none case sensitive)


**`$ grep -i`**

```
import subprocess

search_txt = 'ot'
proc1 = subprocess.Popen(['curl', 'https://www.python-course.eu/simpsons_phone_book.txt'], stdout=subprocess.PIPE)
proc2 = subprocess.Popen(['grep', '-i', search_txt], stdin=proc1.stdout,
                        stdout=subprocess.PIPE, stderr=subprocess.PIPE)

stdout,stderr = proc2.communicate()

print(stdout.decode('utf-8'))
```
