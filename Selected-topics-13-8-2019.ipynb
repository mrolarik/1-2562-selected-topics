```
import math

w = 7   #m
h = 4   #m
Rec_Area = w * h
print(Rec_Area, 'm.2')

a = 3   #cm
a = a/100.0   #m

Squ_Area = math.pow(a,2)
print(Squ_Area, 'm.2')

Area = Rec_Area - Squ_Area
print('Total area', Area, 'm.2')
```

```
Rec1 = 4 * 3
Rec2 = 2.5 * 3.5
Intersec = 0.5 * 0.5

Area = (Rec1 + Rec2) - Intersec
print(Area, 'm.2')
```

```
import math

r1 = 8
r2 = 20
pi = math.pi
area1 = pi * (math.pow(r1,2))
area2 = pi * (math.pow(r2,2))
area = area2 - area1

print('area is', area, 'cm.2')
```

```
import math

r1 = 8
r2 = 20
pi = math.pi
c1 = 2 * pi * r1
c2 = 2 * pi * r2
circumference = c2 - c1

print('c1=', c1)
print('c2=', c2)
print('Difference', circumference, 'cm')
```
